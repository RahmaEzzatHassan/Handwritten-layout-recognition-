```python
from ultralytics import YOLO
import dill

# Load the fine-tuned YOLO model
model_path = '/kaggle/input/yolo88/yolov8x-doclaynet-epoch64-imgsz640-initiallr1e-4-finallr1e-5 (1).pt'
model = YOLO(model_path)

# Retrieve and print the class names and number of classes
class_names = model.names  # Retrieve class names from the model
num_classes = len(class_names)  # Number of classes

print(f"Number of classes: {num_classes}")
print("Class names:")
for i, name in class_names.items():
    print(f"{i}: {name}")
```

```python
from ultralytics import YOLO
import os
import yaml
import torch
'''
# Define the labels and dataset paths
original_labels = {
    'Caption': 0,
    'Footnote': 1,
    'Formula': 2,
    'List-item': 3,
    'Page-footer': 4,
    'Page-header': 5,
    'Picture': 6,
    'Section-header': 7,
    'Table': 8,
    'Text': 9,
    'Title': 10
}
'''
# Your labels
new_labels = {
    'Figure': 0,
    'Headline': 1,
    'None': 2,
    'Subtitle': 3
}

# Creating a merged label dictionary
#labels = {**original_labels, **{k: len(original_labels) + v for k, v in new_labels.items()}}

train_path = '/kaggle/input/fine-tune/Dataset/images/train'
val_path = '/kaggle/input/fine-tune/Dataset/images/val'

# Create the YAML configuration for the dataset
data_dict = {
    'train': train_path,
    'val': val_path,
    'nc': len(new_labels),
    'names': list(new_labels.keys())
}

yaml_path = 'your_data.yaml'
with open(yaml_path, 'w') as f:
    yaml.dump(data_dict, f)

print(f"Dataset configuration saved to {yaml_path}")

# Load the pre-trained YOLOv8 detection model
model = YOLO('/kaggle/input/yolo88/yolov8x-doclaynet-epoch64-imgsz640-initiallr1e-4-finallr1e-5 (1).pt')

# Fine-tune the model
model.train(data='your_data.yaml', epochs=70, batch=14, imgsz=704, lr0=0.001)
```
